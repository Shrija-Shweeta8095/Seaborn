## Seaborn barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

1. x

```python
sns.barplot(x=tips_df.total_bill)
```

2. x,y

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill)
```

3. hue

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill,hue=tips_df.sex)
```

4. data

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df)
```

5. order

```python
order=['Sun','Tue','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order)
```

6. hue_order

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,hue_order=hue_order)
```

7. estimator

```python
import numpy as np
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,estimator=np.mean)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,estimator=np.max)
```

8. cr- confidence interval

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,ci=12)
```

9. n_boot- to count the ci

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,estimator=np.max,n_boot=2)
```

10 units- identifier for sampling unit

11. orient- horizontal or vertical

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,orient='v')
```

```python
sns.barplot(y='day',x='total_bill',hue='sex',data=tips_df) # to plot vertically
```

12. color

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,color='g')
```

13. palette

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,palette='hot')
```

14. saturation

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,saturation=0.3)
```

15. errcolor

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,errcolor='0.5')
```

16. increase the width of error

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,errwidth=12)
```

17. capsize

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,capsize=1)
```

18. dodge- combine both bar in one

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df, dodge=False)
```

19. alpha, linestyle,linewidth,edgecolor in kwargs

```python
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,**kwargs)
```

```python
sns.set()
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,**kwargs)
```

```python
sns.set()
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',data=tips_df,**kwargs)
```

```python

sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,alpha=.9,linestyle="-.",linewidth=3,
            edgecolor='g')
```

```python

ax=sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,alpha=.9,linestyle="-.",linewidth=3,
            edgecolor='g')
ax.set(title="Bar plot of tips dataframe",
      xlabel="Days",
      ylabel="Total Bill")
```
