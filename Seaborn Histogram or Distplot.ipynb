## Histogram or distplot

```python
import seaborn as sns
```

```python
tips_df=sns.load_dataset("tips")
```

```python
tips_df
```

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"])
```

2. bins

```python
sns.distplot(tips_df["total_bill"],bins=100)
```

3. hist=True

```python
sns.distplot(tips_df["total_bill"],bins=100,hist=False)
```

4. kde by default true

```python
sns.distplot(tips_df["total_bill"],bins=100,kde=False)
```

5. rug- means rug plot by default False

```python
sns.distplot(tips_df["total_bill"],bins=100,rug=True)
```

fit 

```python
from scipy.stats import norm
```

```python
sns.distplot(tips_df["total_bill"],fit=norm)
```

```python
sns.distplot(tips_df["total_bill"],fit=norm,kde=False)
```

color

```python
sns.distplot(tips_df["total_bill"],color='r')
```

vertical by default False

```python
sns.distplot(tips_df["total_bill"],color='r',vertical=True)
```

norm_hist by default False value

```python
sns.distplot(tips_df["total_bill"],norm_hist=True)
```

axlabel

```python
sns.distplot(tips_df["total_bill"],axlabel="Tota bill")
```

label

```python
import matplotlib.pyplot as plt
sns.distplot(tips_df["total_bill"],label="Total Bills")
plt.legend()
```

```python
sns.distplot(tips_df["total_bill"],label="Total Bills")
plt.title("Histogram of Total Bill")
plt.legend()
```

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"],label="Total Bills")
plt.title("Histogram of Total Bill")
plt.legend()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins)
plt.title("Histogram of Total Bill")
plt.xticks(bins)
plt.show()
```

hist_kws

```python
sns.set()
sns.distplot(tips_df["total_bill"],hist_kws={'color':'r'})
```

edgecolor, linewidth

```python
sns.set()
sns.distplot(tips_df["total_bill"],hist_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9})
```

kde_kws- kernel density estimator

```python
sns.set()
sns.distplot(tips_df["total_bill"],
             hist_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9},
             kde_kws={'color':'g','linewidth':5,'linestyle':'--','alpha':0.9})
```

rug,rug_kws

```python
sns.set()
sns.distplot(tips_df["total_bill"],
             hist_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9},
             kde_kws={'color':'g','linewidth':5,'linestyle':'--','alpha':0.9},
            rug=True,
            rug_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9})
```

fit_kws

```python
sns.set()
sns.distplot(tips_df["total_bill"],
             hist_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9},
             kde=False,
             fit=norm,
             fit_kws={'color':'g','linewidth':12,'linestyle':'--','alpha':0.4},
             kde_kws={'color':'g','linewidth':5,'linestyle':'--','alpha':0.9},
            rug=True,
            rug_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9})
```

```python
sns.set()
sns.distplot(tips_df["total_bill"],
             hist_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9},
             kde=False,
             fit=norm,
             fit_kws={'color':'g','linewidth':12,'linestyle':'--','alpha':0.4},
             kde_kws={'color':'g','linewidth':5,'linestyle':'--','alpha':0.9},
            rug=True,
            rug_kws={'color':'r','edgecolor':'y','linewidth':5,
                                             'linestyle':'--','alpha':0.9},
            label="TB")

plt.title("Histogram of Restaurant Total Bill",fontsize=15)
plt.xlabel("Total Bill",fontsize=15)
plt.legend()
plt.show()
```

combine multiple

```python
sns.distplot(tips_df["total_bill"],bins=9,label="total_bill")
sns.distplot(tips_df["tip"],bins=9,label="tip")
sns.distplot(tips_df["size"],bins=9,label="size")
plt.legend()
```
